# Preprocessing

```python
# Split Independent and Dependent Variable
X = zoo.iloc[:,1:17].values
y = zoo.iloc[:,[17]].values
```

```python
X
```

```python
y
```

```python
# Split the data into train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
# Find n_neighbours using cross validation score
k_range = range(1,31)
k_score = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5)
    k_score.append(scores.mean())
plt.plot(k_range,k_score)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validation Accuracy')
plt.show()
```

# KNN Model

```python
# Create KNN Model 
Classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
```

```python
Classifier.fit(X_train,y_train)
```

```python
# Predict the Test Values using KNN Train Model
y_pred = Classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
# Check Confusion Matrix table
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Check How much Accuracy of our KNN Model
acc = accuracy_score(y_test, y_pred)
acc
```
