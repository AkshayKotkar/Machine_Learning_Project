```python
#Independent Variable
'''Predict one variable with a set of Multiple Variables'''
X = data.iloc[:,3:5].values
```

```python
X # All variable are independent (Matrix)
```

```python
# This is Complete our Data Preprocessing Template
# After This Template Apply in ML Model
# And then Data Visulization
```

```python
# Fitting Model
# optimal no. of cluster using Elbow Method
# Solving this used KMeans and elbow method choose optimal no of cluster
from sklearn.cluster import KMeans
# form sklearn module import cluster module and there KMeans Function
wcss = []
# wcss is within cluster sum of square it is list for stored error data
for i in range(1,11):
# creating loop create 10 iteration, each iteration run 2 things 1)fit the k means algo to our data X
# 2) compute within cluster sum of sqaure and added to our wcss list
    kmeans = KMeans(n_clusters = i, init= 'k-means++', max_iter=300, n_init= 10, random_state= 0)
    '''create kmeans varible and use the KMeans class import from modules
    kmeans several parameters 
    1) n_cluster i i use for 1 to 10 cluster
    2) init its Kmeans++ method it going to use address a random intialization method for centroid used k-means++
    3) max_iteration there can be find Final Cluster which means how many times algo is run 
    4) n_init it is says no.of times algo run with differnt sets of randomly centroid, how many no. of centroid there'''
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    # inertia it is computes within cluster some of squares
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Cluster')
plt.ylabel('Wcss')
plt.show()
# in output elbow method is 5 no. of cluster bcz after 5 error was very slow decreases
# Choose no. of cluster when point which there is very slow decreses will be no of cluster
```

```python
# Fitting K Means  Clustering to the dataset
kmeans = KMeans(n_clusters = 5, init= 'k-means++', max_iter=300, n_init= 10, random_state= 0)
y_hc = kmeans.fit_predict(X)
# tagging and fitting my X Varible taht is independent variable
```

```python
y_hc
# Each data point tagged to a cluster output in 0 to 4  bcz have 5 cluster
# data point 1 belongs to cluster no 4 output like this in 1st cluster more data points
```

```python
# Visulization of cluster using scatter plot 
```

```python
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s=100, c='red', label= 'Cluster 1')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s=100, c='green', label= 'Cluster 2')
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,1], s=100, c='blue', label= 'Cluster 3')
plt.scatter(X[y_hc == 3,0], X[y_hc == 3,1], s=100, c='yellow', label= 'Cluster 4')
plt.scatter(X[y_hc == 4,0], X[y_hc == 4,1], s=100, c='orange', label= 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=50, c= 'black', label= 'Centroid')
plt.title('Cluster of Customers')
plt.xlabel('Anual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
