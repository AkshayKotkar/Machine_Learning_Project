```python
#Independent Variable
'''Predict one variable with a set of Multiple Variables'''
X = data.iloc[:,3:5].values
```

```python
X # All variable are independent (Matrix)
```

```python
# This is Complete our Data Preprocessing Template
# After This Template Apply in ML Model
# And then Data Visulization
```

```python
# Fitting Model
# optimal no. of cluster using dendrogram
import scipy.cluster.hierarchy as sch
#scipy is scientific python package. there is module cluster and there in the another module is hierarchy
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
# in the hierarchy there is fuction called dendrogram function and used linkage metohod is ward 
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Fitting Hierarchical Clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
# use sklearn model in there cluster model and import AgglomerativeClustering function
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
#in AgglomerativeClustering there input a no. of cluster and type of linkage and affinity euclidean distance
y_hc = hc.fit_predict(X)
# tagging and fitting my X Varible taht is independent variable
```

```python
y_hc
# Each data point tagged to a cluster output in 0 to 4  bcz have 5 cluster
# data point 1 belongs to cluster no 4 output like this in 1st cluster more data points
```

```python
# Visulization of cluster using scatter plot 
```

```python
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s=100, c='red', label= 'Cluster 1')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s=100, c='green', label= 'Cluster 2')
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,1], s=100, c='blue', label= 'Cluster 3')
plt.scatter(X[y_hc == 3,0], X[y_hc == 3,1], s=100, c='yellow', label= 'Cluster 4')
plt.scatter(X[y_hc == 4,0], X[y_hc == 4,1], s=100, c='orange', label= 'Cluster 5')
plt.title('Cluster of Customers')
plt.xlabel('Anual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
